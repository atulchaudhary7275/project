```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')

```

```python
df['Day']=df['Date_of_Journey'].dt.day
df['Month']=df['Date_of_Journey'].dt.month
df['Year']=df['Date_of_Journey'].dt.year
```

```python
df.head(2)
```

```python
df.drop(['Date_of_Journey'],axis=1)
```

```python
df['Day']=df['Day'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'])

# Extract minutes and seconds into separate columns
df['Hours'] = df['Dep_Time'].dt.strftime('%H')  
df['Minutes'] = df['Dep_Time'].dt.strftime('%M')  # Extract minutes
df['Seconds'] = df['Dep_Time'].dt.strftime('%S')  # Extract second
```

```python
df.head(3)
```

```python
df['Hours']=df['Hours'].astype(int)
df['Minutes']=df['Minutes'].astype(int)
df['seconds']=df['Seconds'].astype(int)
```

```python
df.drop(['Dep_Time'],axis=1,inplace=True)

```
